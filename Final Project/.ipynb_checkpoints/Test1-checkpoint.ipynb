```python
len(reddit), len(fileDF)
```

```python
def checker(wrong_options,correct_options):
    names_array=[]
    ratio_array=[]    
    for wrong_option in wrong_options:
        if wrong_option in correct_options:
            names_array.append(wrong_option)
            ratio_array.append('100')
        else:   
            x=process.extractOne(wrong_option,correct_options,scorer=fuzz.token_set_ratio)
            names_array.append(x[0])
            ratio_array.append(x[1])
    return names_array,ratio_array
```

```python
str2Match = fileDF['Brewery Name'].fillna('#####').tolist()
strOptions = reddit['Brewery Name'].fillna('#####').tolist()
```

```python
nameMatch, ratioMatch = checker(str2Match, strOptions)

matchDF = pd.DataFrame()
matchDF['Names to Check'] = pd.Series(str2Match)
matchDF['Correct Names'] = pd.Series(nameMatch)
matchDF['Correct Ratio'] = pd.Series(ratioMatch)

matchDF.head()
```

### Fuzzy 2

https://medium.com/analytics-vidhya/matching-messy-pandas-columns-with-fuzzywuzzy-4adda6c7994f

`fuzz.ratio()` compares 2 strings & outputs a ratio

```python
fuzz.ratio("It's sunny outside!", "Outside is really sunny")
```

`process.extract()` finds the best matches in a list of strings

```python
list_of_strings = ['a', 'ab', 'c', 'cd']
process.extract( 'aaa', list_of_strings, limit=2 )
```

#### Different *scorers*:
`fuzz.ratio()`  
`fuzz.partial_ratio()`  
`fuzz.WRatio()`, etc.  
`process.extract(..., scorer=fuzz.ratio)`  
`process.extract(..., scorer=fuzz.partial_ratio)`  
`process.extract(..., scorer=fuzz.WRatio)`, etc.  

##### *process.extract*

```python
newDF = pd.DataFrame()
checkDF = pd.read_csv('nebraskaCleaned.csv')
actualDF = pd.read_csv('testData.csv')
newDF['NamesToCheck'] = checkDF['Brewery Name']
```

```python
actualName = []
similarity = []

for i in newDF.NamesToCheck:
    ratio = process.extract(i, actualDF['Brewery Name'], limit=1)
    actualName.append(ratio[0][0])
    similarity.append(ratio[0][1])

newDF['Actual Name'] = pd.Series(actualName)
newDF['Similarity'] = pd.Series(similarity)
```

```python
newDF.head()
```

```python
newDF.sort_values('Similarity', ascending=False)
```

### Try Other FuzzyWuzzy 'Scorers'

```python
scorer_dict = { 'R':fuzz.ratio, 
                'PR': fuzz.partial_ratio, 
                'TSeR': fuzz.token_set_ratio, 
                'TSoR': fuzz.token_sort_ratio,
                'PTSeR': fuzz.partial_token_set_ratio, 
                'PTSoR': fuzz.partial_token_sort_ratio, 
                'WR': fuzz.WRatio, 
                'QR': fuzz.QRatio,
                'UWR': fuzz.UWRatio, 
                'UQR': fuzz.UQRatio }
```

```python
testDF = pd.DataFrame()
checkDF = pd.read_csv('nebraskaCleaned.csv')
actualDF = pd.read_csv('testData.csv')
testDF['NamesToCheck'] = checkDF['Brewery Name']
```

```python
def scorerTesterFunction(x):
    testDF = pd.DataFrame()
    testDF['NamesToCheck'] = checkDF['Brewery Name']
    
    actualName = []
    similarity = []
    
    for i in testDF['NamesToCheck']:
        ratio = process.extract(i, actualDF['Brewery Name'], limit=1,
                               scorer=scorer_dict[x])
        actualName.append(ratio[0][0])
        similarity.append(ratio[0][1])
        
        testDF[x+'Name'] = pd.Series(actualName)
        testDF[x+'Score'] = pd.Series(similarity)
        
    print('Average Score: {0:.2f}'.format(testDF[x+'Score'].mean()))
        
    return testDF.sort_values(x+'Score', ascending=False)
```
