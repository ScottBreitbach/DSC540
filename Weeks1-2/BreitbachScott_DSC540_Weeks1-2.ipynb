1) Create a **multiline_text** variable by copying the text from the first chapter of *Pride and Prejudice*.

```python
with open("P&P Ch01", "r") as file:
    multiline_text = file.readlines()
    multiline_text = ",".join(multiline_text)
    multiline_text = multiline_text.lower()
```

3) Remove all new lines and symbols using the **replace** function.

```python
special_chars = ["\n", ",", "!", ".", "?", ";", '"', "  ", "  ", "  "]
cleaned_text = multiline_text
for c in special_chars:
    cleaned_text = cleaned_text.replace(c, " ")
```

```python
cleaned_text
```

4) Find all of the words in **multiline_text** using the **split** function.

```python
all_words = cleaned_text.split(" ")
```

```python
all_words
```

5) Create a list from this list that will contain only the unique words.

```python
unique_words = list(set(all_words))
```

```python
unique_words
```

6) Count the number of times the unique word has appeared in the list by using the **key** and **value** in **dict**.

```python
word_dict = {word : all_words.count(word) for word in unique_words}
```

```python
word_dict
```

7) Find the top 25 words from the unique words that you have found using the **slice** function.

```python
sorted_words = sorted(word_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
top_25 = list((dict(sorted_words[0:25])).keys())
```

6) Combine the preceding code into one block, and this time write a separate function where you will pass the list generated from **dropwhile**, and the function will return the whole number contained in the list (**[1, 2]** becomes **12**). Make sure that the return type is a number and not a string. Treat the incoming list as a stack in the function and generate the number by reading the individual digits from the stack.

```python
def ReturnWholeNum(stack):
    collect = ''
    for i in range(0, len(stack)):
        tos = str(stack.pop())
        collect = tos + collect
    collect = int(collect)
    return collect       

for i in permutations(range(3), 3):
    assert isinstance(i, tuple) == True
    i = list(dropwhile(lambda x: x==0, i))
    assert isinstance(i, list) == True
    i = ReturnWholeNum(i)
    print(i)
```
