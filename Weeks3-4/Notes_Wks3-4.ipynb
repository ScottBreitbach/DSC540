### Reindexing

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
```

Calling **reindex** rearranges the data according to the new index:

```python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])
obj2
```

Use **ffill** while reindexing to forward-fill the values:

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index = [0, 2, 4])
obj3
```

```python
obj3.reindex(range(6), method='ffill')
```

In a DF, **reindex** can alter index (row), coluns, or both.  
When passed only a sequence, it reindexes the rows:

```python
frame =  pd.DataFrame(np.arange(9).reshape((3, 3)),
                     index=['a', 'c', 'd'],
                     columns=['Ohio', 'Texas', 'California'])
frame
```

```python
frame2 = frame.reindex(['a', 'b', 'c', 'd'])
frame2
```

Columns can be reindexed with the **columns** keyword:

```python
states = ['Texas', 'Utah', 'California']
frame.reindex(columns=states)
```

### Function Application and Mapping

NumPy ufuncs (element-wise array methods) also work with pandas objects:

```python
frame = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

Get absolute value:

```python
np.abs(frame)
```

Apply a function on 1-D arrays to each column or row using DF's **apply** method:

```python
f = lambda x: x.max() - x.min()
```

```python
frame.apply(f)  # by column
    # result is a Series with the columns of 'frame' as its index
```

If you pass **axis='columns'** to **apply**, the function will be per row instead:

```python
frame.apply(f, axis='columns')
```

Don't need to use **apply** for methods that already exist in DF (like **sum** and **mean**).

The function passed to **apply** can return a Series w/multiple values (instead of a scalar value):

```python
def f(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
frame.apply(f)
```

Element-wise functions can be used too.  
You could compute a formatted string from each floating-point value in 'frame' with **applymap**:

```python
format = lambda x: '%.2f' % x
```

```python
frame.applymap(format)
```

```python
frame['e'].map(format)
```

```python
frame.rank(axis='columns')
```
